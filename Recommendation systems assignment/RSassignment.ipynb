```python
# prompt: how to change the name of column in df

df.rename(columns={
    'User.ID': 'User_Id'
}, inplace=True)
```

```python
len(df.User_Id.unique())
```

```python
df = df.drop_duplicates(subset='User_Id', keep='first')
```

```python
user_df = df.pivot(index='User_Id',
                                 columns='Book.Title',
                                 values='Book.Rating')
```

```python
user_df
```

```python
#Impute those NaNs with 0 values
user_df.fillna(0, inplace=True)

user_df
```

```python
from sklearn.metrics import pairwise_distances
user_sim = 1 - pairwise_distances(user_df.values,metric='cosine')
```

```python
user_sim
```

```python
# diagnoal are replaced with zero
import numpy as np
np.fill_diagonal(user_sim, 0)
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
user_sim_df
```

```python
#Set the index and column names to user ids
user_sim_df.index   = df.User_Id.unique()
user_sim_df.columns = df.User_Id.unique()
user_sim_df
```
